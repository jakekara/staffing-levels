
# Add ECS grant to staffing and fiscal dataframe

Now we're going to add the combined data from 3 and 4, and then add wealth ranks, to produce:

    output/5_combined.csv

```python
import pandas as pd
```

```python
df = pd.read_csv("output/3_staffing-with-fiscal.csv").set_index(["City","report_year"])
df.head()
```

```python
grant_spending = pd.read_csv("output/4_grant_spending.csv").set_index(["City","report_year"])
grant_spending.columns = ["Grant Payments"]
grant_spending.head()
```

```python
df = df.join(grant_spending,how="outer")
df.head()
```

## Add wealth rank

```python
wealth = pd.read_excel("data/aen2019_Excel.xls",
                       skiprows=17,
                       header=None)

wealth = wealth[[1,6]]

wealth.columns = ["City","wealth_rank"]

wealth["City"] = wealth["City"].apply(lambda x: str(x).strip())

wealth = wealth.set_index("City")

wealth
```

```python
df = pd.merge(df.reset_index(), wealth.reset_index(), on="City").set_index(["City","report_year"])
```

```python
df[df["wealth_rank"].isnull()]
```

```python
df.to_csv("output/5_combined.csv")
```
