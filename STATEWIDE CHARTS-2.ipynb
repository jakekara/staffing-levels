```python
import matplotlib.pyplot as plt
import numpy as np
def dfScatter(df, xcol='Height', ycol='Weight', catcol='Gender'):
    fig, ax = plt.subplots()
    categories = np.unique(df[catcol])
    colors = np.linspace(0, 1, len(categories))
    print colors
    colordict = dict(zip(categories, colors))
    print ("categories", categories)

    df["Color"] = df[catcol].apply(lambda x: colordict[x])
    ax.scatter(df[xcol], df[ycol], c=df.Color)
    return fig

def updown(val, interval=500):
    try:
        return str(int((val) / interval) * interval)
    except:
        pass
    
print updown(2300, 500)

combined["updown"] = combined["grants_per_student_change"].apply(updown)
dfScatter(combined[(combined["report_year"] == 2017)
                   &(combined["staffing_per_student_change"] < -0.2)], xcol="wealth_rank", ycol="grants_per_student_change", catcol="updown")
```
